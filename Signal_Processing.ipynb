## What is a Signal
What is a Signal?

A signal is a way to represent information using numbers.

In audio:

- X-axis → Time

- Y-axis → Amplitude (loudness / air pressure)

Types of Signals

- Analog signal: Continuous (real-world sound)

- Digital signal: Discrete (computer-understandable)

## Frequency
What is Frequency?

Frequency tells us how many times a signal repeats in one second.

- Unit: Hertz (Hz)

- Low frequency → bass sound

- High frequency → sharp sound

Example:

- 5 Hz → 5 cycles per second

- 440 Hz → musical “A” note

Frequency = pitch of sound

## Sine Wave (Basic Signal)
What is a Sine Wave?

A sine wave is the simplest audio signal.

#### Formula:

#### x(t)=Asin(2πft)
Where:

- A → Amplitude (loudness)

- f → Frequency (Hz)

- t → Time

Why sine wave first?

- Any complex sound = combination of sine waves

- Easy to understand frequency behavior

## Windowing
Why Windowing?

Audio is processed in short frames.
Frame edges are sharp → causes spectral leakage.

## Hamming Window
What is Hamming Window?

A window function that:

- Hamming window is used to smoothly reduce the signal values at the edges of a frame so that frequency leakage is reduced.

Formula:



- w[n] --> This is the window value at position n, It tells how much weight the signal sample at position n
- 0.54 (Constant term) --> This sets the overall height of the window , Ensures the middle of the window is close to 1
- 0.46 (Constant term) --> Controls how much the edges are reduced, Larger value → stronger tapering at edges
- cos() --> Creates a smooth curve, Avoids sudden jumps at edges
- 2π --> Represents one full cycle of cosine, Standard for periodic signals
- n --> Position of the sample inside the frame, Starts from 0 to N-1
- N (Frame length)

Why Hamming?

- Best balance for speech processing

## Power Spectrum
What is Power Spectrum?

Power spectrum tells how much energy is present at each frequency of the signal.

Formula:



| Term   | Meaning                    |
| ------ | -------------------------- |
| (P(k)) | Power at frequency bin k   |
| (X(k)) | FFT output                 |         
| (N)    | FFT length (normalization) |


Why power?

- Represents energy

- Used for Mel filtering

## Filtering
What is Filtering?

Removing unwanted frequencies.

Types:

- Low-pass → removes high noise

- High-pass → removes low noise

- Band-pass → keeps speech band

- 300 Hz – 3400 Hz
